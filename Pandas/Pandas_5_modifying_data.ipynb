# Modifying Data - Updating rows and columns

```python
import pandas as pd
people = {
    "first" : ["Devraj","Yash", "Tarun"],
    "last" : ["Charan", "Burad", "Yadav"],
    "email" : ["devrajcharan@gmail.com", "yashburad@gmail.com", "tarunyadav@gmail.com"]
}
df = pd.DataFrame(people)
df
```

```python
df.columns
```

```python
#to rename column]
df.columns = ['first name', 'last name', 'email']
df
```

```python
#to upper case all columns
df.columns = [x.lower() for x in df.columns]
df
```

```python
#to remove spaces or anything we use string methods
df.columns = df.columns.str.replace(' ', '_')
df
```

```python
#to change particular columns, use dictionary
df.rename(columns= {'first_name': 'first',
                    'last_name' : 'last'})
#but it just create a copy, do not change names of original DataFrame
```

```python
df
```

```python
#to change for original use, inplace = True
df.rename(columns= {'first_name': 'first', 'last_name' : 'last'},inplace = True)
df
```

## For rows

```python
#for particular row
df.loc[2] = ['Tarun','Yadav','yadavtarun@gmail.com']
df
```

```python
#for specific element
df.loc[2,'email'] = 'tarunyadav@gmail.com'
df
#you can do this for multiple columns to by passing a list
# df.loc[0,['col_1',''col_2' ]] = ['new_col1',''new_col2']
```

```python
#but its prefered to use df.at when modifying specifir=c value do avoid overhead from loc/iloc
df.at[0,'email'] = 'charandevraj@gmail.com'
df
```

```python
#you can use filter to modify more efficiently
filt = (df['email'] == 'charandevraj@gmail.com')
df.loc[filt,'email'] = 'devrajcharan@gmail.com'
df
```

```python
#to apply sting methods on paticular value
df['email'] = df['email'].str.upper()
df
```

```python
df.loc[[0,2],'email'] = df.loc[[0,2],'email'].str.lower()
df
```

## Methods to modify multiple rows
## 1. apply
## 2. map
## 3. applymap
## 4. replace

```python
#apply - To apply any function on whole series
df['email'].apply(len)
```

```python
#you can also use user created function ->
def update_email(email):
    return email.lower()
df['email'].apply(update_email)
```

```python
df['email'] = df['email'].apply(update_email)
df
```

```python
#you can also apply lemda function ->
df['email'] = df['email'].apply(lambda email: email.upper())
df
```

```python
#applying on whole DataFrame
df.apply(len) #applies funtion on each series
```

```python
df.apply(len, axis = 'columns')
```

```python
df.apply(min)
```

```python
df.apply(min, axis = 1)
```

```python
# applymap - to apply any function on each element of a DataFrame
df.applymap(len) #it is removed now, no longer works in newer version of pandas
#instead just use df.map()
```

```python
df
```

```python
df['email'] = df['email'].apply(str.lower)
df
```

```python
# map - only works on series, used for substituing each value in a series
df['first'].map({'Devraj': 'Dev', 'Yash': 'Yeshu'})
# We get NaN for not provided value, to avoid that use 'replace' method
```

```python
#replace - to replace particular values of a series
df['first'].replace({'Devraj': 'Dev', 'Yash': 'Yeshu'})
df #this won't update the original df 
```

```python
#replace - to replace particular values of a series
modify = df['first'].replace({'Devraj': 'Dev', 'Yash': 'Yeshu'})
modify
```

```python
df['first'] = modify
df
```
