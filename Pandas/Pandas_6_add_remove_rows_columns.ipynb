## Add/Remove rows and columns

```python
import pandas as pd
people = {
    "first" : ["Devraj","Yash", "Tarun"],
    "last" : ["Charan", "Burad", "Yadav"],
    "email" : ["devrajcharan@gmail.com", "yashburad@gmail.com", "tarunyadav@gmail.com"]
}
df = pd.DataFrame(people)
df
```

## For columns

```python
# TO add columns -> df['col1'] + ' ' + df['col2']
df['first'] + ' ' + df['last']
```

```python
#To make a new column for first and last
df['full_name'] = df['first'] + ' ' + df['last']
df
```

```python
#To remove columns -> df.drop(columns = [<columns names>])
df.drop(columns = ['first', 'last']) #won't change original
```

```python
df
```

```python
df.drop(columns = ['first', 'last'], inplace = True)
df
```

```python
# to split a column-> use string split()
df['full_name'].str.split(' ')
```

```python
# to split columns into columns -> after split( ,expand=True)
df['full_name'].str.split(' ', expand = True)
```

```python
#To add these columns in DataFrame
df[['first','last']] = df['full_name'].str.split(' ', expand = True)
df.drop(columns = ['full_name'], inplace = True)
df
```

## For rows

```python
#To add row
first = {'first' : ['Aryan']}
df1 = pd.DataFrame(first)
df_new = pd.concat([df, df1])
df_new
```

```python
people2 = {
    "first" : ["Pradyumn","Aryan"],
    "last" : ["Krishna", "Sharma"],
    "email" : ["pradyumnkrishna@gmail.com", "aryansharma@gmail.com"]
}
df2 = pd.DataFrame(people2)
df2
```

```python
#to add rows of one DataFrame to another
df = pd.concat([df,df2], ignore_index = True)
df
```

```python
#to remove rows -> df.drop(index = index of row)
df.drop(index = 3)
```

```python
df.drop(index = 3, inplace = True)
```

```python
df
```

```python
#you can also apply filter while using drop
filt = df['last'] == 'Sharma'
df.drop(index = df[filt].index)
```
