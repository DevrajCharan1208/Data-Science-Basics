## Filtering

```python
import pandas as pd
people = {
    "first" : ["Devraj","Yash", "Tarun"],
    "last" : ["Charan", "Burad", "Yadav"],
    "email" : ["devrajcharan@gmail.com", "yashburad@gmail.com", "tarunyadav@gmail.com"]
}
df = pd.DataFrame(people)
df
```

```python
filt = df['first'] == 'Devraj'
df[filt]
#you can also
df[df['first']=='Devraj'] #without assigning variable
```

```python
df.loc[filt]
```

```python
#you can also get specific column by using df.loc
df.loc[filt, 'email']
```

```python
filt = (df['last'] == 'Charan') & (df['first'] == 'Devraj')
df.loc[filt]
```

```python
filt = (df['first'] == 'Devraj') | (df['last'] == 'Yadav')
df.loc[filt, 'email']
```

```python
#you can also use '~' for negation of filter
df.loc[~filt, 'email']
```

```python
# for multiple filters, 
# 1.) make a list
first = ['Devraj','Yash']
# 2.) us df['column'].isin(list_name)
filt = df['first'].isin(first)
df.loc[filt]
```

## String Method

```python
#To check is the string contains specific word
filt = df['email'].str.contains('devraj')
df.loc[filt,['first','last']]
```
