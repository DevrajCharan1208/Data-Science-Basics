## Filtering operation from Pandas_4

```python
df1.columns
```

```python
high_experience = (df['WorkExp'] >= 50 )
df_exp = df.loc[high_experience, ['ResponseId','WorkExp' ,'EdLevel','Country','LanguageWantToWorkWith']]
df_exp
```

```python
#you can also filter out multiple , let's say you want from specific countries
countries = ['United States of America', 'France', 'Germany']
filt = df_exp['Country'].isin(countries)
```

```python
df_exp.loc[filt, 'Country']
```

```python
#String method
filt = df_exp['LanguageWantToWorkWith'].str.contains('Python', na = False) #(na = false) is must to avoid error in NAN values
df_exp.loc[filt,['Country','LanguageWantToWorkWith']]
```
