## Indexing, and operations

```python
import pandas as pd
# dictionary = { key : value }
people = {
    "first" : ["Devraj","Yash", "Tarun"],
    "last" : ["Charan", "Burad", "Yadav"],
    "email" : ["devrajcharan@gmail.com", "yashburad@gmail.com", "tarunyadav@gmail.com"]
}
# here keys are columns if we convert into data frame
#to access use keys ->  dictionary[key]
df = pd.DataFrame(people)
df
```

```python
#To set any column as index -> df.set_index('name')
# we can aslo use df = pd.read_csv('filename.csv', index_col = 'col_name')
df.set_index('email')
#but it won't change the index permanently
```

```python
df
```

```python
#To change index permanently -> df.set_index('name', inplace = True)
df.set_index('email', inplace = True)
```

```python
df
```

```python
#To see all index/indices
df.index #we need these to find certain row
```

```python
df.loc['tarunyadav@gmail.com']
```

```python
df.reset_index(inplace = True)
df
```

```python
#to sort indexes -> df.sort_index
df1  = df.set_index('email')
df1.sort_index()
```

```python
#To reverse sort the index
df.sort_index(ascending = False)

```

```python
#to sort permanently
df1.sort_index(inplace = True)
df1
```
