```python
import pandas as pd
```

## Dictionary

```python
# dictionary = { key : value }
people = {
    "first" : ["Devraj","Yash", "Tarun"],
    "last" : ["Charan", "Burad", "Yadav"],
    "email" : ["devrajcharan@gmail.com", "yashburad@gmail.com", "tarunyadav@gmail.com"]
}
# here keys are columns if we convert into data frame
#to access use keys ->  dictionary[key]
people ['email']
```

```python
#To convert it into 'Data Frame'
df1 = pd.DataFrame(people)
df1
```

```python
#To convert it into 'Data Frame'
df1 = pd.DataFrame(people)
df1
```

```python
#to access a single column -> df[column_name]
df1['email']
```

```python
#i single column of a data frame is a series,
#data frame is also k/as containor for multiple series
type(df1['email'])
```

```python
#column of a data frame can also be accessed by -> df.column_name
#using this method can have problem  if column_name is same as one of the attributes/method of data frame
df1.email
```

```python
#to access multiple columns at same time
df1[['first','last']]
```

```python
#to access all column names
df1.columns
```

```python
#to access rows and columns both (by index)
df1.iloc[[0,1], [0,2]]
```

```python
df1
```

```python
#to access rows, columns by label -> df.loc
df1.loc[[0,1],['first', 'email']]
```
