#### Advance Operations with Buisness examples

```python
import numpy as np
import matplotlib.pyplot as plt

# Data structure: [restaurant_id, 2021, 2022, 2023, 2024]
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000],   # Paradise Biryani
    [2, 120000, 140000, 160000, 190000],   # Beijing Bites
    [3, 200000, 230000, 260000, 300000],   # Pizza Hub
    [4, 180000, 210000, 240000, 270000],   # Burger Point
    [5, 160000, 185000, 205000, 230000]    # Chai Point
])
```

```python
print (" === ZOMATO SALES ANALYSIS === ")

print ("\n shape of sales data :\n", sales_data.shape)
print ("\n first three restaurants :\n", sales_data[:3])

# slicing can be used on rows & columns together by -> sales_data[ : , : ], where sales_data[ rows , columns ]
                          
```

```python
# Sales per year
yearly_sales = np.sum(sales_data[:,1:], axis = 0) # axis = 0, represents columns
print("Total Sales per Year", yearly_sales)

#Minimum Sales per Restaurant
min_sales_restau = np.min(sales_data[:,1:], 1) # 1 -> axis = 1, represents rows
print("Minimus Sales per Restaurant: ", min_sales_restau)

#Maximum Sales per Year
max_sales_year = np.max(sales_data[:,1:], 0)
print("Max Sales Per Year: ", max_sales_year)

#Average Sales per Year
avg_sales_year = np.mean(sales_data[:,1:], 0)
print("Average Sales Per Year: ", avg_sales_year)

#Cumilative Sales Per Restau
cumilative_sales_restau = np.cumsum(sales_data[:,1:], 1)
print("Cumilative Sales Per Restaurant: \n", cumilative_sales_restau)
```

```python
#Plotting data
plt.figure(figsize=(8,4))
plt.plot([2021,2022,2023,2024],np.mean(cumilative_sales_restau, 0))
plt.xticks([2021,2022,2023,2024])
plt.title("Average cumilative sales across all restaurants")
plt.xlabel("Year")
plt.ylabel("Cumilative Sales")
plt.grid(True)
plt.show()
```
