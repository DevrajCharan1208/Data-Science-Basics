## Matrix Operations

```python
import numpy as np 
import time

# MATRIX MULTIPLICATION

A = np.array([[1,2,3],
                    [2,3,4],
                    [3,4,5]])

B = np.array([[3,4,1],
                    [2,5,6],
                    [3,1,2]])

print (" == Matrix Multiplication == \n")

print ("A.dot(B): \n",  A.dot(B), "\n")

print("np.dot(A,B): \n", np.dot(A,B), "\n")

print ("A@B: \n", A@B, "\n")

print ("np.matmul(A,B): \n", np.matmul(A,B), "\n")

# INVERSE OF A MATRIX
print ( "\n== Inverse of a Matrix ==")
print("np.linalg.inv(A): \n", np.linalg.inv(B), "\n") #will raise an error if the matrix is singular
```

```python
# Eigen values of a Matrix
print("== Eigen values of a Matrix ==\n")

print("np.linalg.eigvals(A): ", np.linalg.eigvals(A) )


#Matrix raised to the power n
print("\nnp.linalg.matrix_power(A,n): \n", np.linalg.matrix_power(A,2))

```

### Decomposition

```python
A = np.array([[0, 7, 1, 5],
              [5, 9, 2, 1],
              [2, 4, 6, 5],
              [5, 2, 4, 3]])
print(A)

#QR Decomposition
QR = np.linalg.qr(A) # can get both as Q,R = np.linalg.qr(A)
print("\nnp.linalg.qr(A): \n", QR,"\n Q = \n", QR[0],"\n R = \n", QR[1])
# QR[0] for Q, QR[1] for R

```

```python
# Singular Values of A
singular_values = np.linalg.svdvals(A)
print("Singular Values of A: ", singular_values)

# SVD
SVD = np.linalg.svd(A)
print("SVD Decomposition of A: \n", SVD)

```

## Eigen values and vectors

```python
#Eigen values and vector for a square matrix
square_matrix = np.linalg.eig(A) #If the matrix is not a square matrix it will give error
print(square_matrix)


# Norm, Length
print(np.linalg.norm(A)) #for both matrix and vectors
print(np.linalg.matrix_norm(A)) #for matrix
```

```python
#Determinent of array or matrix
np.linalg.det(A)
```

```python
#Rank of a matrix
np.linalg.matrix_rank(A)
```

```python
#trace 
np.trace(A) #for arrays

np.linalg.trace(A) #for matrix
```


## Solving System of Linear Equations

```python
A = np.array([[1,-1,2],
              [2,1,4],
              [4,-1,1]])
B = np.array([2,7,4])
x = np.linalg.solve(A,B,) #for solution
print("Solution for System of equation: ", x)

y = np.allclose(np.dot(A,x), B) #to check, return boolean
print("Is solution correct: ", y)

#solving least square solution
least_sq = np.linalg.lstsq(A,B)
print("Least square solution: ", least_sq)

```

## Other operations

```python
#diagonal
diag = np.diagonal(A)
print("Diagonals :", diag)

print(np.linalg.diagonal(A))

#Transpose 
transpose1 = A.T
transpose2 = np.linalg.matrix_transpose(A)

print(transpose1, "\n\n", transpose2)
```

```python
#covariance of a matrix
cov = np.cov(A)
print(cov)
```
